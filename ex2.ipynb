```python
#Import Libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

#KMeans from cluster for unsupervised ML
from sklearn.cluster import KMeans
```

```python
#loading data
dataset = pd.read_csv('data.csv')
X = dataset.iloc[:500,:].values

X.shape
```

```python
#Applying KMeans Model
from sklearn.cluster import KMeans
```

```python
#in thes ex i want to change n_clusters untell i find best n_clusters num for higher accurancy
ilist = [] #firist i creat empty list to put inertia on it  "small inertia means higher accurancy"
n = 11 
for i in range(1,n): #loop from 1 to 11 to find best inertia
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(X) #fit for applying KMean algorithms and get output 
    ilist.append(kmeans.inertia_) #here i put inertia on list that i creat
    
print('all inertia :', ilist)
```

```python
#here i plot using Elbow graph
#Elboe graph used to show relation between num of clusters and inertia to help me choose best n_clusters
plt.plot(range(1,n), ilist)
plt.title('Elbow') 
plt.xlabel('Num of clusters')
plt.ylabel('Inertias')
plt.show()
```

```python
#here i choose n_clusters = 6
kmeans = KMeans(n_clusters = 6, init = 'k-means++')
y_kmeans = kmeans.fit_predict(X)
```

```python
# Visualising the clusters
#i have 6 clusters from 0 to 5
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 10, c = 'r')#on cluster 0 color = red and size =10
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 10, c = 'b')#on cluster 1 color = blue and siz =10
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 10, c = 'g')#on cluster 2 color = green and size = 10
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 10, c = 'c')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 10, c = 'm')
plt.scatter(X[y_kmeans == 5, 0], X[y_kmeans == 5, 1], s = 10, c = 'y')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 300, c = 'y')#center data showing as points on yellow color "c = 'y'" &on size = 300 "s =300" 
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
#plt.legend()
plt.show()
```
