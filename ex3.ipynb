```python
#importing libraries
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score #silhouette_score used to calculate accurancy like inertia also small num for better accurancy
```

```python
#loading data
from sklearn.datasets import load_iris #working on built irise data
iris = load_iris()
X = iris.data
```

```python
kmean = KMeans(n_clusters= 3 )

kmean.fit(X)

result = kmean.labels_

print('silhouette_score when num of clusters = 3 :\n' ,silhouette_score(X , result))
```

```python
#in thes ex i want to change n_clusters untell i find best n_clusters num for higher accurancy
score = [] #firist i creat empty list to put inertia on it  "small inertia means higher accurancy"

for n in range(2,11):#loop from 2 to 12 to find best inertia
    kmean = KMeans(n_clusters= n )
    kmean.fit(X)
    result = kmean.labels_
    print(n , '    '  , silhouette_score(X , result))
    score.append(silhouette_score(X , result))
    
print('all silhouette_scor :', score)
```

```python
#here i plot using Elbow graph
#Elboe graph used to show relation between num of clusters and inertia to help me choose best n_clusters
plt.plot(range(2,11) , score)
plt.title('Elbow') 
plt.xlabel('Num of clusters')
plt.ylabel('Silhouette_score')
plt.show()
```

```python
kmean = KMeans(n_clusters= 6 ) #i choose best num of n_clusters from elbow graph depending on when Silhouette_score will be very close to each other on graph


    
y_kmeans = kmean.fit_predict(X)
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 10, c = 'r') #on cluster 0 color = red and size =10
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 10, c = 'b') #on cluster 1 color = blue and siz =10
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 10, c = 'g') #on cluster 2 color = green and siz =10
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 10, c = 'c')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 10, c = 'm')
plt.scatter(X[y_kmeans == 5, 0], X[y_kmeans == 5, 1], s = 10, c = 'y')

plt.scatter(kmean.cluster_centers_[:, 0], kmean.cluster_centers_[:, 1], s = 100, c = 'y') #center data showing as points on yellow color "c = 'y'" &on size = 100 "s =100" 
plt.title('Clusters of flowers')
plt.show()
```
